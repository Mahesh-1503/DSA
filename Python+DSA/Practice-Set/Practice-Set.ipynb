#### Type conversion

* Implicit type conversion

```python
# converting int to float
integer_num = 10
float_num = 10.3
new_num = integer_num + float_num
print("Value: ", new_num)
print("Type: ", type(new_num))
```

* Explicit type conversion

```python
num_string = '12'
num_integer =23
print(f"type of num_string: {type(num_string)}")
print(f"type of num_integer: {type(num_integer)}")
num_string = int(num_string)
print(f"type of num_string after conversion: {type(num_string)}")
num_float = float(num_integer)
print(f"type of num_float after conversion: {type(num_float)}")
```

```python
number = int(input('Enter a number: '))
print(f'You entered: {number}')
if number > 0:
    print('Positive number')
elif number < 0:
    print('Negative number')
else:
    print('Zero')
print('This statement always executes')
```

### Python Nested if Statements
* It is possible to include an if statement inside another if statement.

```python
number = 5
# outer if statement
if number >= 0:
    # inner if statement
    if number == 0:
        print('Zero')
    # inner else statement
    else:
        print('Positive number')
# outer else statement
else:
    print('Negative number')
```

#### Python pass Statement
* In Python programming, the pass statement is a null statement which can be used as a placeholder for future code.

* Suppose we have a loop or a function that is not implemented yet, but we want to implement it in the future. In such cases, we can use the pass statement.

```python
## Using pass With Conditional Statement
n = 10
if n > 10:
    pass # Placeholder for future code
print('Hello, World!')
```

```python
n = 10

if n > 10:
    pass  # Placeholder for future code
    # write code later by removing pass

print('Hello')
```

### Python break and continue

* In programming, the break and continue statements are used to alter the flow of loops:

* break exits the loop entirely
* continue skips the current iteration and proceeds to the next one

* Python break Statement


```python
for i in range(6):
    if i == 4:
        break
    print(i)
```

<pre>
if i == 4
    break
</pre>

* terminates the loop when i is equal to 4. Hence, the output doesn't include values after 3

```python
## break Statement with while Loop
i = 0
while i < 5:
    if i == 3:
        break
    print(i)
    i += 1 
```

* Working of continue Statement in Python

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

* skips the current iteration when i is equal to 3, and continues the next iteration. Hence, the output has all the values except 3.

#### continue Statement with while Loop

```python
# Program to print odd numbers from 1 to user-defined number
limit = int(input("Enter the limit: "))
print("Odd numbers from 1 to", limit, "are:")
num = 0
while num < limit:
    num += 1
    if num % 2 == 0:
        continue
    print(num)
```
